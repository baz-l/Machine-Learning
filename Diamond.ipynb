```python
dummy=pd.get_dummies(df[['Cut','Color','Clarity','Polish', 'Symmetry']],drop_first=True)
dummy
```

```python
df1=pd.concat([df,dummy],axis=1)
df1
```

```python
df1=df1.drop(['Cut','Color','Clarity','Polish', 'Symmetry'],axis=1)
df1
```

```python
x=df1.drop("Price",axis=1)
x
```

```python
y=df1["Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
print(f"The Mean Absolute Error is : {mean_absolute_error(y_test,y_pred)}")
```

```python
print(f"R 2 score is :{r2_score(y_pred,y_test)}")
```
